### Correcting Date Format

```python
#Importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df1 = pd.read_excel("temp.xlsx",sheet_name = 'Sheet1')
df2 = pd.read_excel("temp.xlsx",sheet_name = 'Sheet2')
```

```python
import datetime
```

```python
for i in range(0,len(df1)):
    try:
        df1['DATE'][i]= df1['DATE'][i].strftime('%d/%m/%Y')
    except:
        continue
```

```python
df1.dropna(inplace=True)
df1
```

```python
df1.info()
```

```python
for i in range(0,len(df2)):
    try:
        df2['DATE'][i]= df2['DATE'][i].strftime('%d/%m/%Y')
    except:
        continue
```

```python
df2.dropna(inplace=True)
df2
```

```python
df2.info()
```

```python
df1.set_index('DATE',inplace=True)
df1
```

```python
df2.set_index('DATE',inplace=True)
df2
```

```python
df = pd.concat([df1,df2],axis=1, join='inner')
```
