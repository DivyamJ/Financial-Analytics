### Dynamic Linkages of stock prices between the BRICs and the United States: Effects of 2008-09 Financial Crisis 

```python
#Importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
import pandas_datareader.data as web
import datetime
```

```python
start = datetime.datetime(2004,5,2)
end = datetime.datetime(2020,5,2)
```

```python
us = web.DataReader('^DJI','yahoo',start,end)
brazil = web.DataReader('^BVSP','yahoo',start,end)
russia = pd.read_csv("RTSI-dailyhistory.csv", index_col='Date', parse_dates =True, dayfirst=True)
india = web.DataReader('^BSESN','yahoo',start,end)
china = web.DataReader('000001.SS','yahoo',start,end)
south_africa = web.DataReader('EZA','yahoo',start,end)
```

```python
russia.head()
russia.index
```

```python
china.head()
```

```python
south_africa.head()
```

```python
'''us.to_csv('US Stock Prices.csv')
brazil.to_csv('Brazil Stock Prices.csv')
russia.to_csv('Russia Stock Prices.csv')
india.to_csv('India Stock Prices.csv')
china.to_csv('China Stock Prices.csv')
south_africa.to_csv('South Africa Stock Prices.csv')'''
```

```python
us['US'] = np.log(us['Adj Close']/us['Adj Close'].shift(1))
brazil['Brazil'] = np.log(brazil['Adj Close']/brazil['Adj Close'].shift(1))
russia['Russia'] = np.log(russia['Close']/russia['Close'].shift(1))
india['India'] = np.log(india['Adj Close']/india['Adj Close'].shift(1))
china['China'] = np.log(china['Adj Close']/china['Adj Close'].shift(1))
south_africa['South Africa'] = np.log(south_africa['Adj Close']/south_africa['Adj Close'].shift(1))
```

```python
south_africa.head()
```

```python
russia.head()
```

```python
brazil.head()
```

```python
df = pd.concat([us['US'],brazil['Brazil'],russia['Russia'],india['India'],china['China'],south_africa['South Africa']],axis=1,join='inner')
```

```python
df.dropna(inplace=True)
df.head()
```

```python
df['GFC']=-1
df['Covid']=-1
```

```python
df.head()
```

```python
#Date stock market fell more than 6% in light of GFC - 29th Sep, 2008
#Date first case of coronavirus was detected in China - 17th Nov, 2019

df['GFC']['2004-05-06':'2008-09-28'] = 0
df['GFC']['2008-09-29':'2012-05-01'] = 1
df['Covid']['2012-05-02':'2019-11-16'] = 0
df['Covid']['2019-11-17':'2020-05-02'] = 1
df.tail()
```

```python
df.info()
```

```python
df_pre_gfc = df[df['GFC']==0]
df_pre_gfc.tail()
```

```python
df_post_gfc = df[df['GFC']==1]
df_post_gfc.head()
```

```python
df_pre_covid = df[df['Covid']==0]
df_pre_covid.tail()
```

```python
df_post_covid = df[df['Covid']==1]
df_post_covid.head()
```

```python
#Saving the table
df.to_csv('Final return table.csv')
```

### COINTEGRATION

```python
from statsmodels.tsa.vector_ar.vecm import coint_johansen
```

```python
result = coint_johansen(pd.concat([us['Adj Close'],brazil['Adj Close']], axis=1, join='inner'), -1, 4)
result.lr1
```

```python
result.cvt
```

US and Brazil are not cointegrated

```python
result = coint_johansen(pd.concat([us['Adj Close'],russia['Close']], axis=1, join='inner'), -1, 4)
result.lr1
```

```python
result.cvt
```

US and Russia are not cointegrated

```python
result = coint_johansen(pd.concat([us['Adj Close'],india['Adj Close']], axis=1, join='inner'), -1, 4)
result.lr1
```

```python
result.cvt
```

US and India are weakly cointegrated

```python
result = coint_johansen(pd.concat([us['Adj Close'],china['Adj Close']], axis=1, join='inner'), -1, 4)
result.lr1
```

```python
result.cvt
```

US and China are not cointegrated

```python
result = coint_johansen(pd.concat([us['Adj Close'],south_africa['Adj Close']], axis=1, join='inner'), -1, 4)
result.lr1
```

```python
result.cvt
```

US and South Africa are not cointegrated

```python
result = coint_johansen(pd.concat([india['Adj Close'],china['Adj Close']], axis=1, join='inner'), -1, 4)
result.lr1
```

```python
result.cvt
```

India and China are not cointegrated

#### All nations stock prices are not cointegrated with each other as per the above results except for US and India which is weakly cointegrated
