# Autoregressive modelling

```python
#Importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Importing the dataset
df = pd.read_excel('Lecture 1_Data.xlsx')
df.head()
```

```python
import statsmodels.api as sm
```

```python
d1 = pd.concat([df['GDP'].shift(1),df['GDP'].shift(2),df['GDP'].shift(3),df['GDP'],],axis=1)
d1 = d1.drop([0,1,2])
d1.columns = ['GDP-1','GDP-2','GDP-3','GDP']
d1.head()
```

```python
d1 = sm.add_constant(d1)
d1.head()
```

```python
#Autoregressive of order 3
model1 = sm.OLS(d1.iloc[:,4],d1.iloc[:,:-1]).fit()
model1.summary()
```

```python
#Autoregressive of order 2
model1 = sm.OLS(d1.iloc[:,4],d1.iloc[:,[0,1,2]]).fit()
model1.summary()
```

```python
#Autoregressive of order 1
model1 = sm.OLS(d1.iloc[:,4],d1.iloc[:,[0,1]]).fit()
model1.summary()
```

```python
d2 = pd.concat([df['Prime Rate'].shift(1),df['Prime Rate'].shift(2),df['Prime Rate'].shift(3),df['Prime Rate'],],axis=1)
d2 = d2.drop([0,1,2])
d2.columns = ['Prime Rate-1','Prime Rate-2','Prime Rate-3','Prime Rate']
d2.head()
```

```python
d2 = sm.add_constant(d2)
d2.head()
```

```python
#Autoregressive of order 3
model2 = sm.OLS(d2.iloc[:,4],d2.iloc[:,:-1]).fit()
model2.summary()
```

```python
#Autoregressive of order 2
model2 = sm.OLS(d2.iloc[:,4],d2.iloc[:,[0,1,2]]).fit()
model2.summary()
```

```python
#Autoregressive of order 1
model2 = sm.OLS(d2.iloc[:,4],d2.iloc[:,[0,1]]).fit()
model2.summary()
```
