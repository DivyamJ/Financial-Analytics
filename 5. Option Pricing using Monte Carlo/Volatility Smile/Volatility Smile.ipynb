# Volatility Smiles

```python
#Importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Assumed values
T = 22/365
nT = 500
delT = T/nT
rf = 0.0623
S0= 11251
nsim = 100
tol = 0.1
step_size = 0.005
```

```python
#Reading Option data
df = pd.read_csv("Option Prices.csv")
df.head()
```

```python
df['Implied Volatility'] = 0
df.head()
```

```python
#Monte Carlo with European Options

for i in range(len(df)):
    
    K = df.iloc[i,1]
    sig = df.iloc[i,3]
    price = -100
    
    while(abs(df.iloc[i,2]-price)>tol*df.iloc[i,2]):
        if(sig<0.60):
            sig = sig + step_size
        else:
            sig = 0
        summ = 0 
        for j in range(0,nsim):
            St = S0

            for k in range(0,nT):
                St = St*np.exp((rf-sig**2/2)*delT + sig*np.sqrt(delT)*np.random.randn(1))
            if (df.iloc[i,0]=='Call'):
                payoff = max(0,St-K)
            else:
                payoff = max(0,K-St)
            summ = summ + payoff

        payoff_avg = summ/nsim
        price = payoff_avg*np.exp(-rf*T)
        print('price {} : {}'.format(i,price))
        
    df.iloc[i,3] = sig
    
```

```python
a = df.groupby('Strike Price').mean()
a
```

```python
a.iloc[:,1].plot(title='Volatility Smile')
plt.ylabel('Implied Volatility');
```
